<h2>A note about interactivity</h2>

<p>**Most general-purpose programming environments are non-interactive:**
<ol>
<li>Write code
<li>Start program, execute code
<li>Stop program
<li>Repeat
</ol>

<p>Important reasons for this:
<ul>
<li>Simplicity - interactivity requires complex infrastructure
<li>Repeatability - program must run the same way every time
</ul>

<p>**Most scientific programming environments (Igor, Matlab, Mathematica, ...) prefer an interactive experience.**<br>
Why? Data analysis is incremental.

<ul>
<li>You want to see your data, test some code, check the results, tweak the code, check again, etc.
<li>It would be a pain (and possibly very slow) to have to stop and restart the entire program for every change.
</ul>

<p>Interactivity means:

<ul>
<li>The program that runs your code does not exit when it finishes running your code.
<li>All of your data remains in memory for you to inspect
<li>Code can be modified and executed in the running process.
<li>Maybe nice GUI tools that let you visualize, save, load data without writing any code.
</ul>

<p>**Python supports both types of environment.** The tools we will introduce fall into one category or the other, but all of them incorporate features from both categories.
</div>

<div style="border-left: 3px solid #000; padding: 1px; padding-left: 10px; background: #F0FAFF; ">
<h2>Editor + console</h2>

This is the simplest approach to code development. Use a nice text editor (sublime is a good one) that at least handles indentation. Save your code to a file, and run it from a command line like:

    > python myfile.py
    
This approach falls into the "non-interactive" category, but there are many tools available to make it more interactive. For example, simply running python with the `-i` flag will cause it to begin an interactive session immediately after your script completes:

    > python -i myfile.py
    
This allows you to inspect the variables generated by your program and experiment with new ideas.


<p>**Pros:**
<ul>
<li>Simple, bare-metal approach. Nice if you want to use python without other user interfaces getting in your way. For example:
<ul>
<li>IDEs introduce useful features at the cost of added complexity. (How much time will you spend configuring your IDE instead of coding?) 
<li>Some IDEs have debuggers that slow down your code.
<li>Some IDEs change the way your code runs "under the hood".
</ul>
<li>You decide exactly which tools you want to use.
</ul>

<p>**Cons:**
<ul>
<li>May be more difficult if you're not familiar with DOS/Bash/other CLI
<li>Non-integrated; you need to manually incorporate many other tools (eg for visualization, version control, debugging, etc.)
</ul>
</div>

<div style="border-left: 3px solid #000; padding: 1px; padding-left: 10px; background: #F0FAFF; ">
<h2>Jupyter (IPython) Notebook</h2>

<p>Jupyter Notebook provides an *interactive* programming environment similar to the Mathematica notebook.
<ul>
<li>Code is divided into chunks called "cells".
<li>Cells can be executed in any order, independently of each other.
<li>Results from cell execution are displayed below each cell.
</ul>

A Python interpreter runs *continuously* in the background. Every time you execute a cell, the state of the interpreter is updated.


<p>**Pros:**
<ul>
<li>Notebooks are excellent for storing and presenting the results of data analysis, simulations, etc. because code, results, and discussion are all combined in a single document.
<li>Any user should be able to execute your notebook and reproduce your results.
  (this is true of all code, but not necessarily as easy)
<li>Good for prototyping--easy to break up the script into multiple intermediate steps and iteratively develop a small chunk at a time.
</ul>

<p>**Cons:**
<ul>
<li>Notebooks are not good for developing large applications or user interfaces.
<li>Difficult to use if your project incorporates multiple Python files.
<li>Possible to generate unreproducible results because cells can be executed out of order.
<li>Awkward web browser interface, text editing is clumsy with limited configurability.
<li>Difficult to integrate with debugging and version integration tools.
</ul>
</div>

<div style="border-left: 3px solid #000; padding: 1px; padding-left: 10px; background: #F0FAFF; ">
<h2>PyCharm (IDE)</h2>

<p>PyCharm is a popular integrated development environment. It includes a fully-featured text editor, excellent debugging tools, interactive data browsing, project management, and version control integration. It is easy to install but *not* included with Anaconda.

<p>**Pros:**
<ul>
<li>Excellent, fully-featured text editor.
    <ul>
    <li>Syntax hilighting, indentation, autocomplete
    <li>Context-sensitive help (ctrl-q)
    <li>Can view Jupyter Notebook files.
    </ul>
<li>Excellent debugging tools
    <ul>
    <li>Breakpoints / steppping
    <li>Stack inspection
    <li>Interactive object inspection (plotting with matplotlib!)
    <li>Array visualization
    </ul>
<li>Project management, code versioning
    <ul>
    <li>Tracks files and execution environment
    <li>If you are working on a project with multiple files, this can be very helpful
    </ul>
</ul>

<p>**Cons:**
<ul>
<li>**Lots** of features to wade through, overhead in learning new tools.
<li>Project management can get in the way if you just want to work with a single file.
<li>Interactivity is somewhat limited.
</ul>
</div>

<div style="border-left: 3px solid #000; padding: 1px; padding-left: 10px; background: #F0FAFF; ">
<h2>Spyder</h2>

Spyder is an integrated development environment with many similarities to MATLAB's default environment. It is included with Anaconda and provides a more interactive approach to development. 

Matlab + Notebook - like features:
* Divide script into cells, execute in persistent process
* Inline matplotlib results

<p>**Pros:**
<ul>
<li>Decent text editor, debugger, and project management.
<li>More interactive than pycharm--console, variable explorer, visualization tools.
<li>List of set scripts, good for running already-written analysis tools. 
<li>Notebook-like features--divide a script into cells and execute in a background Python process.
</ul>

<p>**Cons:**
<ul>
<li>Somewhat less mature and polished than pycharm.
<li>Debugger can be difficult to use, no stack inspection.
</ul>
</div>
